```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funcin para generar p(w) para el kernel RBF
def rbf_distribution(w, sigma):
    """Distribucin gaussiana p(w) asociada al kernel RBF."""
    return (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (w / sigma)**2)

# Grfica en 1D para p(w)
def plot_p_w_1d(sigma=1.0):
    """Visualiza la distribucin p(w) en 1D."""
    w = np.linspace(-5 * sigma, 5 * sigma, 500)
    p_w = rbf_distribution(w, sigma)
    
    plt.figure(figsize=(8, 5))
    plt.plot(w, p_w, label=f"p(w), ={sigma}")
    plt.title("Distribucin p(w) para el kernel RBF (1D)")
    plt.xlabel("Frecuencia ")
    plt.ylabel("p(w)")
    plt.grid(True)
    plt.legend()
    plt.show()

# Grfica en 2D para p(w)
def plot_p_w_2d(sigma=1.0):
    """Visualiza la distribucin p(w) en 2D."""
    w1, w2 = np.meshgrid(np.linspace(-5 * sigma, 5 * sigma, 100), 
                         np.linspace(-5 * sigma, 5 * sigma, 100))
    p_w = rbf_distribution(np.sqrt(w1**2 + w2**2), sigma)
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(w1, w2, p_w, cmap='viridis', alpha=0.8)
    ax.set_title("Distribucin p(w) para el kernel RBF (2D)")
    ax.set_xlabel("1")
    ax.set_ylabel("2")
    ax.set_zlabel("p(w)")
    plt.show()

# Generar muestras  para visualizacin en 2D
def sample_w_from_rbf(sigma=1.0, n_samples=500):
    """Genera muestras de  desde p(w) y visualiza."""
    mean = np.array([0, 0])
    cov = sigma**2 * np.eye(2)  # Matriz de covarianza diagonal
    samples = np.random.multivariate_normal(mean, cov, size=n_samples)
    
    plt.figure(figsize=(8, 6))
    plt.scatter(samples[:, 0], samples[:, 1], alpha=0.6, label=f" ~ N(0, {sigma**2}I)")
    plt.title("Muestras  desde p(w) para kernel RBF")
    plt.xlabel("1")
    plt.ylabel("2")
    plt.grid(True)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Ejecucin de las funciones
sigma = 1.0
plot_p_w_1d(sigma)       # Visualizacin de p(w) en 1D
plot_p_w_2d(sigma)       # Visualizacin de p(w) en 2D
sample_w_from_rbf(sigma) # Muestras  desde p(w) para kernel RBF
```
